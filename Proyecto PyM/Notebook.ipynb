# Avance 4

```python
sucursal_mayor_ventas_2da_Enero = """SELECT 
    Sucursal, 
    SUM(Total_Ticket) AS Total_Vendido
FROM 
    Ventas
WHERE 
    Fecha BETWEEN '2025-01-16' AND '2025-01-31'
GROUP BY 
    Sucursal
ORDER BY 
    Total_Vendido DESC
LIMIT 1;"""
t.consultar(sucursal_mayor_ventas_2da_Enero)
```

```python
producto_menos_vendido = """
SELECT 
    Producto,
    SUM(Cantidad_Vendida) AS Total_Cantidad
FROM 
    Ventas
WHERE 
    Fecha BETWEEN '2025-01-01' AND '2025-01-31'
GROUP BY 
    Producto
ORDER BY 
    Total_Cantidad ASC
LIMIT 1;
"""
t.consultar(producto_menos_vendido)
```

```python
sucursal_menos_venta = """
SELECT 
    Sucursal,
    SUM(Cantidad_Vendida) AS Cantidad
FROM 
    Ventas
WHERE 
    Producto = 'Tijeras' AND
    Fecha BETWEEN '2025-01-01' AND '2025-01-31'
GROUP BY 
    Sucursal
ORDER BY 
    Cantidad ASC
LIMIT 1;"""

t.consultar(sucursal_menos_venta)
```

```python
ticket_promedio_0106_Polanco = """
SELECT 
    AVG(Total_Ticket) AS Ticket_Promedio
FROM 
    Ventas
WHERE 
    Fecha = '2025-01-06' AND
    Sucursal = 'Polanco';
"""
t.consultar(ticket_promedio_0106_Polanco)
```
